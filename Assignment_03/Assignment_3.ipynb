```python
import pandas as pd
import numpy as np
from itertools import combinations
```

```python
df = pd.read_csv('moviereviews.csv')   #read csv
```

```python
review = df["review"].to_numpy()
review
```

```python
rev = [ele.split(',') for ele in review]
rev
```

```python
words = []
n = len(rev)
for i in range(n-1):
        words.append(rev[i][0].split())
        
words                                    # rev matrix splits into words matrix
```

```python
df = pd.DataFrame(words)   #convert the words into dataframe
df
```

```python
similarity_matrix = np.zeros((len(words), len(words)))    #define a null matrix of same size equal to number of words

for i, j in combinations(range(len(words)), 2):   # 2 means iter over first and its next row
    set_i = set(df.iloc[i])                       
    set_j = set(df.iloc[j])
    jaccard_sim = len(set_i.intersection(set_j)) / len(set_i.union(set_j))  # jaccard similarity give how much similar the sets are to each other
    similarity_matrix[i, j] = jaccard_sim
    similarity_matrix[j, i] = jaccard_sim

similarity_matrix
```

```python

# Set a threshold for similarity, change it to get level of similarity (lower is better)
similarity_threshold = 0.05       

# Print the similar lists for each list
for i in range(len(words)):
    similar_indices = np.where(similarity_matrix[i] > similarity_threshold)[0]
    if len(similar_indices) > 0:
        print(f"List {i}: Similar Lists: {similar_indices}")
        print("Original List:")
        print(words[i])
        print("Similar Lists:")
        for idx in similar_indices:
           print(words[idx])
        print("\n")
```
